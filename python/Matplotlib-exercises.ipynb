```python
# Exercise 1
import numpy as np
import matplotlib.pyplot as plt

products = ['A', 'B', 'C', 'D']
sales = [150, 200, 100, 250]

fig, ax = plt.subplots()
ax.bar(products, sales, color=['hotpink','purple','green','red'])

plt.title('Sales by product')
plt.xlabel('Products')
plt.ylabel('Sales')
          
plt.show()

# The product with the highest sales is D and the product with the lowest sales is C.
```

```python
# Exercise 2
import numpy as np
import matplotlib.pyplot as plt

hours = [1,2,3,4,5,6,7,8]
grades = [50, 60, 65, 70,80,85,90,95]

plt.title('Relationship of student hours and grades')
plt.xlabel('Hours')
plt.ylabel('Grades')

plt.scatter(hours, grades, color='purple')
          
plt.show()

# The more hours a student dedicates to study, the better grade he/she gets.
```
